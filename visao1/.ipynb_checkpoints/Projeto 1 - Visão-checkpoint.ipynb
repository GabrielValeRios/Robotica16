```python
rintin_gray.shape
```

```python
plt.imshow(rintin_gray, cmap="Greys_r")
```

```python

hist,bins = np.histogram(rintin_gray.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
imagem = cv2.imread("hall_box_battery1.jpg")
plt.imshow(imagem)
```

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
imagem_rgb_t = imagem_rgb.transpose()
imagem_rgb_t
```

```python
imagem_r = imagem_rgb[:,:,0]
imagem_g = imagem_rgb[:,:,1]
imagem_b = imagem_rgb[:,:,2]

imagem_r_nova = imagem_r-imagem_g
imagem_b_nova = imagem_b-imagem_g
```

```python
plt.imshow(imagem_r_nova, cmap = 'Greys_r')
imagem_r_nova
```

```python

hist,bins = np.histogram(imagem_r_nova.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(imagem_r_nova.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
plt.imshow(imagem_b_nova, cmap='Greys_r')
imagem_b_nova
```

```python
#valores de i retirados do histograma, onde possivelmente nossa imagem esta

def acha_caixa_vermelha(i):
    if i > 100 and i < 180:
        return 255
    else:
        return 0

#transformar em array 
    
array_acha_caixa_vermelha = np.vectorize(acha_caixa_vermelha)

caixa_vermelha = [array_acha_caixa_vermelha(i) for i in imagem_r_nova]
caixa_vermelha = np.asarray(caixa_vermelha)

plt.imshow(caixa_vermelha, cmap = "Greys_r")
```

```python
def acha_bateria_azul(i):
    if i > 30 and i < 80:
        return 255
    else:
        return 0
    
array_acha_bateria_azul = np.vectorize(acha_bateria_azul)

bateria_azul = [array_acha_bateria_azul(i) for i in imagem_b_nova]

bateria_azul = np.asarray(bateria_azul)

plt.imshow(bateria_azul, cmap = "Greys_r")
```

```python
imagem_caixa_final = caixa_vermelha - bateria_azul
```
